```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('~/Desktop/COVID19/data/200913COVID19MEXICO.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
df['RESULTADO'].unique()
# Out: array([1, 2, 3])

# Filtrar por casos positivos.
# 1 = CASOS POSITIVOS
df = df[df['RESULTADO'] == 1]
df.drop(['RESULTADO'], axis = 1, inplace = True)

drop_cols = ['FECHA_ACTUALIZACION', 'ID_REGISTRO', 'SECTOR', 'ENTIDAD_UM', 'INTUBADO', 'NEUMONIA', 'UCI',
             'SECTOR', 'ORIGEN', 'ENTIDAD_NAC', 'ENTIDAD_RES', 'MUNICIPIO_RES', 'NACIONALIDAD',
             'HABLA_LENGUA_INDIG', 'OTRO_CASO', 'MIGRANTE', 'PAIS_NACIONALIDAD', 'PAIS_ORIGEN', 
            'FECHA_INGRESO', 'FECHA_SINTOMAS', 'FECHA_DEF', 'EMBARAZO']
df.drop(drop_cols, axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
## VARIABLE RESPUESTA = TIPO_PACIENTE
## - 1: Ambulatorio
## - 2: Hospitalizado

n_amb, n_hos = df['TIPO_PACIENTE'].value_counts()
print("Total de casos positivos: {:,}".format(len(df)))
print("Número de casos ambulatorios: {:,}".format(n_amb))
print("Número de casos hospitalizados: {:,}".format(n_hos))
print("El {:.1f}% de los casos positivos son hospitalizados.".format(100*n_hos/(n_amb+n_hos)))
```

```python
# CORRELACIÓN DE TIPO_PACIENTE CON LAS DEMÁS COLUMNAS
print(df.corr()['TIPO_PACIENTE'])
```

```python
## VALORES ÚNICOS DE CADA COLUMNA
## Al parecer 98 y 97 son nulls
for col in df.columns:
    print(col)
    print("nunique:", df[col].nunique())
    print("values:\n",df[col].unique())
    print("")
```

```python
## Cambiamos todos los 98 a ceros excepto en la columna 'EDAD'.
cols = list(df.columns)
cols.remove('EDAD')
for col in cols:
    df[col] = df[col].replace(98, 0)

## Verificamos que ya no haya 98 en las columnas que no sean 'EDAD'.
for col in df.columns:
    print(col)
    print("nunique:", df[col].nunique())
    print("values:\n",df[col].unique())
    print("")
```
